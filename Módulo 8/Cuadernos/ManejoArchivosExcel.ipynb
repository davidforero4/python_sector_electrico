```python
import openpyxl

wb = openpyxl.load_workbook('ejemplo2.xlsx')
```

```python
ws = wb.active
```

```python
print('Número de filas: '+str(ws.max_row)+'. Cantidad de columnas: '+str(ws.max_column))
```

```python
datos_tabla=[ws.cell(row=i,column=3).value for i in range(2,7)]
print(datos_tabla)
```

```python
# Para rango tanto de filas como de columnas:

datos_tabla = [[ws.cell(row=i, column=j).value for j in range(2, 5)]
    for i in range(2, 7)]
```

```python
datos_tabla
```

```python
df = pd.DataFrame(datos_tabla)

print(df)
```
