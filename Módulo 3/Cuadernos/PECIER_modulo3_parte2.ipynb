## Ejemplos

**Ejemplo 3.6:** Cantidad indefinida de parámetros (Avanzado)

```python
def promedio_consumo(*consumos):
    if not consumos:
        return 0
    return sum(consumos) / len(consumos)

print(promedio_consumo(120, 150, 135, 140))  # 136.25
```

**Ejemplo 3.7:**

```python
def crear_reporte(**datos):
    print("\n--- REPORTE ---")
    for clave, valor in datos.items():
        print(f"{clave.title().replace('_', ' ')}: {valor}")

crear_reporte(
    equipo="T-125",
    voltaje_operacion=220,
    ubicacion="Planta Sur",
    estado="Operativo"
)
```

**Ejemplo 3.8:**

```python
def analizar_circuito(voltaje, corriente):
    potencia = voltaje * corriente
    resistencia = voltaje / corriente if corriente else float('inf')
    estado = "OK" if potencia < 5000 else "SOBRECARGA"
    return potencia, resistencia, estado

p, r, e = analizar_circuito(220, 15)
print(f"P: {p}W, R: {r:.2f}Ohm, Estado: {e}")
```

**Importante:** Observe que en el llamado a la función se establecen los nombres de las variables de retorno.



**Ejemplo 3.9:**

```python
def diagnostico_motor(voltaje, corriente, rpm):
    potencia_entrada = voltaje * corriente
    potencia_mecanica = rpm * 0.1047 * corriente
    eficiencia = (potencia_mecanica / potencia_entrada) if potencia_entrada else 0
    estado = "NORMAL" if 0.7 <= eficiencia <= 0.9 else "REVISAR"

    return {
        "potencia_entrada": potencia_entrada,
        "potencia_mecanica": potencia_mecanica,
        "eficiencia (%)": round(eficiencia * 100, 2),
        "estado": estado
    }

reporte = diagnostico_motor(220, 8.5, 1750)
for clave, valor in reporte.items():
    print(f"{clave}: {valor}")
```
